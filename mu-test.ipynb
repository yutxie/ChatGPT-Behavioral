```python
# perform Wilcoxon ranksums test: 
from scipy.stats import ranksums

baseline = 43 - 1
print('baseline:', notes[baseline])
for i in range(baseline+1, len(lists)):
    print(i+1, format(ranksums(lists[baseline], lists[i])[1], '.10f'), notes[i])
```
